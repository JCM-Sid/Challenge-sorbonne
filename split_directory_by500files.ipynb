```python
# This script to split train/test data in batches of 500 files
#
#
import torch
import numpy as np
import pandas as pd
import os
import random
import csv
from datetime import datetime

current_directory = os.getcwd()
print(f"The current working directory is: {current_directory}")

training_path_dir =   r"D:\ChallengeDST\folder_training_set\folder_training_set"
test_path_dir =   r"D:\ChallengeDST\folder_test_set\folder_test_set"
train_meta_data =  r'..\..\Data\ChallengeSorbonne\training_set_metadata.csv'

# read Meta CSV input and save in df
df_meta_train = pd.read_csv(train_meta_data, sep=";")

split_char = "\\"
```

```python
# Split training and test directories, by batch of 500 files
#
for option in ["training", "test"]:
    # init
    file_with_err = 0
    count_files = 0
    df_file_not_in_meta = pd.DataFrame(columns=['path'])
    df_file_to_split = pd.DataFrame(columns=['orign path', 'batch'])
    
    if option == "training":
        path_dir = training_path_dir
    else:
        path_dir = test_path_dir

    # working loop
    for root, _, files in os.walk(path_dir):
        print(f"There are {len(files)} files in the directory: {path_dir}")
        new_rep = f"rep_{count_files}"
        print("Step:", count_files, new_rep)

        full_file_list = [os.path.join(root, file) for file in files]
        for full_path_file in full_file_list:
        
            hash_name = full_path_file.split('.jso')[0]
            hash_name = hash_name.split(split_char)[-1]
            count_files += 1

            if count_files % 500 == 0:
                new_rep = f"rep_{count_files}"
                print("Step:", count_files, new_rep)

            if option == "training":
                if hash_name not in df_meta_train['name'].values:
                    file_with_err += 1
                    df_file_not_in_meta = pd.concat([df_file_not_in_meta, pd.DataFrame([{'path': full_path_file}])], ignore_index=True)
                    print(f"Error: File not found in metadata: {full_path_file}")
                    continue
            
            df_file_to_split = pd.concat([df_file_to_split, pd.DataFrame([{'orign path': full_path_file, 'batch': new_rep}])], ignore_index=True)
    
    print(f"Total {option}:{count_files} , Files not found: {file_with_err}")
    print(df_file_to_split.shape)
    print(df_file_to_split.groupby('batch').size())


    df_file_to_split.tail(5)

    if option == "training":
        df_file_to_split.to_csv("split_training.csv", index=False)
    else:
        df_file_to_split.to_csv("split_test.csv", index=False)
    
```
