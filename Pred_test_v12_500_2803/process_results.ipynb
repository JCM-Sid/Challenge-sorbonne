```python
import pandas as pd

#filename_result = r'G:\Mon Drive\Colab Notebooks\Sorbonne-Challenge-v12\test_prediction_part.csv'
#filename_result = r'G:\Mon Drive\Colab Notebooks\Sorbonne-Challenge-v12\test_prediction.csv'
filename_result = r'..\Pred_test_v12_500_2803\test_prediction.csv'


df_results = pd.read_csv(filename_result)
df_results.head()
```

```python
import random

list_columns = df_results.columns
nan_rows_indices = df_results[df_results['64bits'].isna()].index
pred_row = 3000 - len(nan_rows_indices)
print(f"There are {pred_row} hash-graphs already filled with predictions\nneed to complete filling with probability for {len(nan_rows_indices)} hash-graphs " )
```

```python

val_prob = []
# Create a copy of the DataFrame to avoid SettingWithCopyWarning
df_results_copy = df_results.copy()

for idx in range(len(nan_rows_indices)):  # Iterate up to 5 or the number of NaN rows
    row_index = nan_rows_indices[idx]
    val_list = []
    for col_idx in range(1,len(list_columns)):
        col_name = list_columns[col_idx]
        count_col = df_results_copy[col_name].sum() 
        pred_prob = count_col / pred_row if pred_row > 0 else 0
        val_prob.append(pred_prob)
        random_number = random.random()
        if random_number < pred_prob:
            df_results_copy.loc[row_index,col_name] = 1
            val_list.append(1)
        else:
            df_results_copy.loc[row_index,col_name] = 0
            val_list.append(0)

    print(df_results.loc[nan_rows_indices[idx]]['name'], val_list)

df_results = df_results_copy
```
