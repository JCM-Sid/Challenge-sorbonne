```python
import pandas as pd
import os
import os
from google.colab import drive

# for Google Colab only
if 'COLAB_RELEASE_TAG' in os.environ:
    drive.mount('/content/drive', force_remount=True)
```

```python
train_meta_data = '/content/drive/MyDrive/Sorbonne_Data_challenge/training_set_metadata.csv'
file_to_check =   '/content/drive/MyDrive/Colab Notebooks/Sorbonne-Challenge-v12/split_training-colab-v12.csv'
#file_to_check = '/content/drive/MyDrive/Colab Notebooks/Sorbonne-Challenge-v12/split_val-colab-v12.csv'

df_meta_train = pd.read_csv(train_meta_data, sep=";")
df_training_split = pd.read_csv(file_to_check, sep=",")
full_file_list = df_training_split['orign path']
# training, Validation, PredTest
mode = "training"
split_char = "/"
err_file = 0

for full_path_file in full_file_list:

    if os.path.exists(full_path_file):
        file_size = os.path.getsize(full_path_file)
        if file_size/1000 > 200_000: #file too big
            print("file too big ",full_path_file, file_size/1000)
            err_file +=1
    else:
        #print("file not exist ",full_path_file)
        err_file +=1

    if mode in ["training", "validation"]  :
      hash_name = full_path_file.split('.jso')[0]
      hash_name = hash_name.split(split_char)[-1]
      # files not in metadata
      if hash_name not in df_meta_train['name'].values:
          print("not in meta", hash_name, full_path_file)
          err_file +=1

print("Total nb of files checked:", len(full_file_list), "nb of files in error", err_file)
```
